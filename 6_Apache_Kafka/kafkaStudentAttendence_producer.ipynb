```python
from kafka import KafkaProducer
import time

producer = KafkaProducer(bootstrap_servers='localhost:9092')

students = [
    "Alice present",
    "Bob present",
    "Charlie present"
]

for s in students:
    producer.send('attendance-topic', s.encode())
    print("Attendance:", s)
    time.sleep(1)

producer.flush()
```
