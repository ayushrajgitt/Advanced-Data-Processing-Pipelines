```python
from kafka import KafkaProducer
import random
import time

producer = KafkaProducer(bootstrap_servers='localhost:9092')

while True:
    price = random.randint(100,200)
    msg = f"StockPrice:{price}"

    producer.send('stock-topic' , msg.encode())
    print("Stock Update:", msg)

    time.sleep(2)
```
