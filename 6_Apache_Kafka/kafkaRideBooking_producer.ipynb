```python
from kafka import KafkaProducer
import time

producer = KafkaProducer(bootstrap_servers='localhost:9092')

rides = [
    "Ride booked by User1",
    "Driver assigned",
    "Ride started",
    "Ride completed"
]

for r in rides:
    producer.send('ride-topic', r.encode())
    print("Ride Event:", r)
    time.sleep(2)

producer.flush()
```
