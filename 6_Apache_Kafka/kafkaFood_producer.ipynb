```python
from kafka import KafkaProducer
import time

producer = KafkaProducer(bootstrap_servers='localhost:9092')

orders=[
    "Order 1: Pizza",
    "Order 2: Burger",
    "Order 3: Pasta",
    "Order 4: Sandwich",
    "Order 5: IceCream"
]

for order in orders:
    producer.send('food-orders' , order.encode())
    print("New Orders:", order)
    time.sleep(2)

producer.flush()
```
