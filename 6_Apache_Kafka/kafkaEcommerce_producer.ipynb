```python
from kafka import KafkaProducer
import time

producer = KafkaProducer(bootstrap_servers='localhost:9092')

orders = [
    "OrderID:101 Product:Laptop",
    "OrderId:102 Product:Phone",
    "OrderID:103 Product:Headphones",
    "OrderID:104 Product:charger"
]

for order in orders:
    producer.send('orders-topic', order.encode())
    print("Order sent:", order)
    time.sleep(2)

producer.flush()
```
