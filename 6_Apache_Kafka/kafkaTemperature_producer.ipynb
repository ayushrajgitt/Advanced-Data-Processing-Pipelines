```python
from kafka import KafkaProducer
import random
import time

producer = KafkaProducer(bootstrap_servers='localhost:9092')

while True:
    temperature = random.randint(20,40)
    msg = f"Temperature:{temperature}"
    
    producer.send('sensor-topic', msg.encode())
    print("Sent:", msg)
    
    time.sleep(3)
```
