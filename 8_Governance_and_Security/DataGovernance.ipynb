Data Governance ensures that data is accurate, traceable, secure, and well-documented

```python
# ==========================================
# STEP 1 - IMPORTS
# ==========================================
import pandas as pd
from datetime import datetime

# ==========================================
# STEP 2 - METADATA
# ==========================================
metadata = {
    "table_name": "sales data",
    "owner": "data_engineering_team",
    "created_date": str(datetime.now()),
    "source": "csv_file",
    "description": "E-commerce sales data"
}

# ==========================================
# STEP 3 - LINEAGE TRACKER
# ==========================================
lineage = []

def log_step(step):
    lineage.append(step)
    print(f"[LINEAGE] {step}")

# ==========================================
# STEP 4 - LOAD DATA
# ==========================================
def load_data():
    log_step("Data Loaded")

    df = pd.DataFrame({
        "id":[1,2,2,4],
        "name":["Amit",None,"Riya","John"],
        "sales":[50000,-20000,20000,None]
    })

    print("\nRaw Data:\n", df)
    return df

# ==========================================
# STEP 5 - DATA QUALITY CHECK
# ==========================================
def validate_data(df):
    log_step("Data Validation")

    report = {
        "null_values": df.isnull().sum().to_dict(),
        "duplicates": int(df.duplicated().sum()),
        "negative_values": int((df["sales"] < 0).sum())
    }

    print("\nValidation Report:\n", report)
    return df, report

# ==========================================
# STEP 6 - CLEAN DATA
# ==========================================
def clean_data(df):
    log_step("Data Cleaning")

    df = df.drop_duplicates()
    df["name"].fillna("Unknown", inplace=True)
    df["sales"] = df["sales"].apply(lambda x: abs(x) if pd.notnull(x) else x)
    df = df.dropna()

    print("\nClean Data:\n", df)
    return df

# ==========================================
# STEP 7 - TRANSFORM DATA
# ==========================================
def transform_data(df):
    log_step("Data Transformation")

    df["tax"] = df["sales"] * 0.1
    df["total"] = df["sales"] + df["tax"]

    print("\nTransformed Data:\n", df)
    return df

# ==========================================
# STEP 8 - GENERATE REPORT
# ==========================================
def generate_report(df):
    log_step("Report Generated")

    summary = df.groupby("name")["total"].sum()

    print("\nAnalysis Report:\n", summary)
    return summary

# ==========================================
# STEP 9 - GOVERNANCE REPORT
# ==========================================
def governance_report(validation_report):
    print("\n========== GOVERNANCE REPORT ==========")

    print("\nMetadata:")
    for k, v in metadata.items():
        print(f"{k}: {v}")

    print("\nLineage:")
    for step in lineage:
        print("->", step)

    print("\nData Quality Report:")
    for k, v in validation_report.items():
        print(f"{k}: {v}")

# ==========================================
# STEP 10 - PIPELINE
# ==========================================
def pipeline():
    df = load_data()
    df, validation_report = validate_data(df)
    df = clean_data(df)
    df = transform_data(df)
    generate_report(df)
    governance_report(validation_report)

# ==========================================
# RUN PIPELINE
# ==========================================
pipeline()
```
