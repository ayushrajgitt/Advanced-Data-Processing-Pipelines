```python
import pandas as pd
from datetime import datetime

df = pd.DataFrame({
    "id": [1,2,2,4],
    "name":["Amit", None, "Riya", "John"],
    "sales":[50000, -20000, 20000, None]
})

df
```

```python
metadata = {
    "table_name": "sales_data",
    "owner": "data_engineering_team",
    "created_date": str(datetime.now()),
    "source": "csv_file"
}

metadata
```

Metadata is data about data. It describes information like source, structure, and ownership of a dataset.

```python
lineage = [
    "Data Loaded",
    "Data Validated",
    "Data Cleaned",
    "Data Transformed",
    "Report Generated"
]

lineage
```

Data lineage tracks the flow of data from source to final output, showing how it is processed and transformed.

```python
validation_report = {
    "null_values": df.isnull().sum().to_dict(),
    "duplicates": int(df.duplicated().sum()),
    "negative_values": int((df["sales"]<0).sum())
}

validation_report
```

```python
df = df.drop_duplicates()

df["name"]=df["name"].fillna("Unknown")
df["sales"]=df["sales"].apply(lambda x: abs(x) if pd.notnull(x) else x)
df=df[df["sales"].notnull()]

df
```

```python
df["tax"]=df["sales"]*0.1
df["total"]=df["sales"]+df["tax"]

df
```

```python
result=df.groupby("name")["total"].sum()
result
```
