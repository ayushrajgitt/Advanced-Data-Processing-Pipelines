```python
import pandas as pd

df = pd.DataFrame({
    "id":[1,2,3],
    "name":["Amit","Riya","John"],
    "email":["amit@gmail.com","riya@gmail.com","john@gmail.com"], #sensitive
    "salary":[50000,60000,55000],   #sensitive
    "account_number":["1234567890","9876543210","1122334455"]   # sensitive
})

df
```

```python
masked_df = df.copy()  # can be human readable

masked_df["email"] = masked_df["email"].apply(lambda x: x[0] + "***@***.com")
masked_df["salary"] = "***MASKED***"
masked_df["account_number"] = masked_df["account_number"].apply(lambda x: "XXXXXX" + x[-4:])

masked_df
```

```python
def access_data(role):
    if role == "admin":
        print("Admin Access")
        return df

    elif role == "analyst":
        print("Analyst Access")
        return masked_df

    else:
        print("Access Denied")
        return None
```

```python
from cryptography.fernet import Fernet

key = Fernet.generate_key()
cipher = Fernet(key)

message = "Salary = 50000"

encrypted = cipher.encrypt(message.encode())
decrypted = cipher.decrypt(encrypted).decode()

print("Original:", message)
print("Encrypted:", encrypted)
print("Decrypted:", decrypted)
```

```python
def secure_pipeline(role):
    data = access_data(role)

    if data is not None:
        print("Processing data...")
        print(data.head())
    else:
        print("No access → cannot process")
```

```python
secure_pipeline("analyst")
```

```python
secure_pipeline("admin")
```

```python
secure_pipeline("intern")
```
