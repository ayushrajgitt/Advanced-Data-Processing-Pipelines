```python
# ======================================
# STEP 1 – IMPORTS
# ======================================
import pandas as pd
from cryptography.fernet import Fernet

# ======================================
# STEP 2 – CREATE SENSITIVE DATA
# ======================================
def create_data():
    print("\n[STEP] Creating Sensitive Data...\n")

    df = pd.DataFrame({
        "id":[1,2,3],
        "name":["Amit","Riya","John"],
        "email":["amit@gmail.com","riya@gmail.com","john@gmail.com"],
        "salary":[50000,60000,55000],
        "account_number":["1234567890","9876543210","1122334455"]
    })

    print(df)
    return df

# ======================================
# STEP 3 – DATA MASKING
# ======================================
def mask_data(df):
    print("\n[STEP] Masking Sensitive Data...\n")

    masked_df = df.copy()

    masked_df["email"] = masked_df["email"].apply(lambda x: x[0] + "***@***.com")
    masked_df["salary"] = "***MASKED***"
    masked_df["account_number"] = masked_df["account_number"].apply(lambda x: "XXXXXX" + x[-4:])

    print(masked_df)
    return masked_df

# ======================================
# STEP 4 – IAM (ROLE-BASED ACCESS)
# ======================================
def access_data(role, df, masked_df):
    print(f"\n[STEP] Checking Access for role: {role}\n")

    if role == "admin":
        print("Full Access Granted")
        return df

    elif role == "analyst":
        print("Limited Access (Masked Data)")
        return masked_df

    else:
        print("Access Denied")
        return None

# ======================================
# STEP 5 – ENCRYPTION
# ======================================
def encrypt_data(text):
    print("\n[STEP] Encrypting Data...\n")

    key = Fernet.generate_key()
    cipher = Fernet(key)

    encrypted = cipher.encrypt(text.encode())

    print("Encrypted:", encrypted)

    return encrypted, key

# ======================================
# STEP 6 – DECRYPTION
# ======================================
def decrypt_data(encrypted, key):
    print("\n[STEP] Decrypting Data...\n")

    cipher = Fernet(key)
    decrypted = cipher.decrypt(encrypted).decode()

    print("Decrypted:", decrypted)
    return decrypted

# ======================================
# STEP 7 – SECURE PIPELINE
# ======================================
def secure_pipeline(role):
    print("\n========== SECURE PIPELINE START ==========\n")

    df = create_data()
    masked_df = mask_data(df)

    data = access_data(role, df, masked_df)

    if data is not None:
        print("\nProcessing Data...\n")
        print(data.head())

        # Encrypt one sensitive field
        encrypted, key = encrypt_data("Salary=50000")

        # Decrypt it back
        decrypt_data(encrypted, key)

        print("\nPipeline Completed Successfully\n")
    else:
        print("\nPipeline Stopped Due to Unauthorized Access\n")

    print("\n========== PIPELINE END ==========\n")

# ======================================
# RUN PIPELINE
# ======================================
secure_pipeline("admin")
```
