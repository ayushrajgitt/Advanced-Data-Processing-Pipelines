```python
import pandas as pd
import time

# STEP 1 - Load Data
def load_data():
    print("\n[LOAD]")

    df = pd.DataFrame({
        "id": [1,2,2,3],
        "name": ["Amit", None, "Riya", "John"],
        "sales": [50000, -20000, 20000, None]
    })

    print(df)
    return df

# STEP 2 - Validate Data
def validate_data(df):
    print("\n[VALIDATE]")

    print("Null Values:\n", df.isnull().sum())
    print("Duplicates:", df.duplicated().sum())

    return df

# STEP 3 - Clean Data
def clean_data(df):
    print("\n[CLEAN]")

    df = df.drop_duplicates()
    df["name"] = df["name"].fillna("Unknown")
    df["sales"] = df["sales"].abs()
    df = df[df["sales"].notnull()]

    print(df)
    return df

# STEP 4 - Transform Data
def transform_data(df):
    print("\n[TRANSFORM]")

    df["tax"] = df["sales"] * 0.1
    df["total"] = df["sales"] + df["tax"]

    print(df)
    return df

# STEP 5 - Analyze Data
def analyze_data(df):
    print("\n[ANALYZE]")

    result = df.groupby("name")["total"].sum()
    print(result)

    return result

# STEP 6 - Save Data
def save_data(df):
    print("\n[SAVE]")

    df.to_csv("output.csv", index=False)
    print("Saved as output.csv")

# MAIN PIPELINE
def pipeline():
    df = load_data()
    df = validate_data(df)
    df = clean_data(df)
    df = transform_data(df)
    analyze_data(df)
    save_data(df)

# SCHEDULER
for i in range(2):
    print(f"\n--- Run {i+1} ---")
    pipeline()
    time.sleep(2)
```
