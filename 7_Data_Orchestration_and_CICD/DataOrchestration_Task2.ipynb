📝 Question
Create a pipeline for sales processing:
Steps:
1️⃣ Load dataset
2️⃣ Validate data
3️⃣ Clean data
4️⃣ Analyze total sales
💻 Requirements
✔ Use pandas
✔ Create small dataset
✔ Add at least one data issue (null/negative)
✔ Fix it

Hunt

df.isnull()
df.fillna()
df.groupby()

```python
import pandas as pd

# STEP 1 - Load Data
def load_data():
    print("Loading Data")

    df = pd.DataFrame({
        "name": ["Amit", "Riya", None],
        "sales": [50000, -20000, 30000]
    })

    print(df)
    return df

# STEP 2 - Validate Data
def validate_data(df):
    print("\nValidating Data")

    print("Null Values:\n", df.isnull().sum())
    print("Negative Sales:\n", (df["sales"] < 0).sum())

    return df

# STEP 3 - Clean Data
def clean_data(df):
    print("\nCleaning Data")

    df["name"] = df["name"].fillna("Unknown")
    df["sales"] = df["sales"].abs()

    print(df)
    return df

# STEP 4 - Analyze Data
def analyze_data(df):
    print("\nAnalyzing Data")

    result = df.groupby("name")["sales"].sum()
    print(result)

# PIPELINE
df = load_data()
df = validate_data(df)
df = clean_data(df)
analyze_data(df)
```
