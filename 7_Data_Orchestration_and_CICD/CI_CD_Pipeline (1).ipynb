```python
import pandas as pd

df = pd.DataFrame({
    "order_id":[1,2,2,4,5],
    "customer":["Amit",None,"Riya","John","Sara"],
    "email":["amit@gmail.com","riya@gmail","riya@gmail.com","john@gmail.com","sara@gmail.com"],
    "product":["Laptop","Phone","Phone","Tablet",None],
    "amount":[50000,-20000,20000,None,30000],
    "account_number":["1234567890","9876543210","1111222233","4444555566","7777888899"]
})

df
```

```python
df.drop_duplicates(inplace=True)
df["customer"].fillna("Unknown", inplace=True)
df["amount"] = df["amount"].apply(lambda x: abs(x) if pd.notnull(x) else x)
df = df.dropna()
```

```python
df["tax"] = df["amount"] * 0.1
df["total"] = df["amount"] + df["tax"]
```

```python
metadata = {
    "table_name": "sales_data",
    "owner": "data_team"
}

lineage = ["loaded", "cleaned", "transformed"]
```

```python
masked_df = df.copy()
masked_df["email"] = masked_df["email"].apply(lambda x: x[0] + "***@***.com")
masked_df["account_number"] = masked_df["account_number"].apply(lambda x: "XXXXXX" + x[-4:])
```

```python
def access(role):
    if role == "admin":
        return df
    elif role == "analyst":
        return masked_df
    else:
        return None
```

```python
def pipeline(role):
    data = access(role)

    if data is not None:
        print("Processing...")
        print(data.head())
    else:
        print("Access denied")
```

```python
pipeline("admin")
pipeline("analyst")
pipeline("intern")
```

```python
def ci_cd_pipeline(role):
    print("CI: Validating Data...")

    # CI step
    if df.isnull().sum().sum() > 0:
        print("Fixing missing values...")

    print("CD: Running Pipeline...")

    # CD step
    pipeline(role)
```
