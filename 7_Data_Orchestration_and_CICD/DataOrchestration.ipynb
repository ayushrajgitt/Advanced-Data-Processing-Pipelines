```python
# ==========================================
# STEP 1 - IMPORTS
# ==========================================
import pandas as pd
import time

# ==========================================
# STEP 2 - LOGGER FUNCTION
# ==========================================
def log(step):
    print(f"[LOG] {step} started...")
    time.sleep(1)

# ==========================================
# STEP 3 - TASK 1: LOAD DATA
# ==========================================
def load_data():
    log("Loading Data")

    data = {
        "id":[1,2,2,4],
        "name":["Amit",None,"Riya","John"],
        "sales":[50000,-20000,20000,None]
    }

    df = pd.DataFrame(data)
    print(df)

    return df

# ==========================================
# STEP 4 - TASK 2: DATA QUALITY CHECK
# ==========================================
def validate_data(df):
    log("Validating Data")

    print("Null Values:\n", df.isnull().sum())
    print("Duplicates:\n", df.duplicated().sum())
    print("Negative Sales:\n", (df["sales"] < 0).sum())

    return df

# ==========================================
# STEP 5 - TASK 3: CLEAN DATA
# ==========================================
def clean_data(df):
    log("Cleaning Data")

    df = df.drop_duplicates()
    df["name"].fillna("Unknown", inplace=True)
    df["sales"] = df["sales"].apply(lambda x: abs(x) if pd.notnull(x) else x)
    df = df.dropna()

    print(df)

    return df

# ==========================================
# STEP 6 - TASK 4: TRANSFORM DATA
# ==========================================
def transform_data(df):
    log("Transforming Data")

    df["tax"] = df["sales"] * 0.1
    df["total"] = df["sales"] + df["tax"]

    print(df)

    return df

# ==========================================
# STEP 7 - TASK 5: ANALYZE DATA
# ==========================================
def analyze_data(df):
    log("Analyzing Data")

    result = df.groupby("name")["total"].sum()

    print("Analysis Result:\n", result)

    return result

# ==========================================
# STEP 8 - TASK 6: SAVE DATA
# ==========================================
def save_data(df):
    log("Saving Data")

    df.to_csv("final_output.csv", index=False)
    print("Data saved as final_output.csv")

# ==========================================
# STEP 9 - PIPELINE (DAG)
# ==========================================
def pipeline():
    print("\n===== PIPELINE STARTED =====\n")

    try:
        df = load_data()
        df = validate_data(df)
        df = clean_data(df)
        df = transform_data(df)
        result = analyze_data(df)
        save_data(df)

        print("\n===== PIPELINE COMPLETED SUCCESSFULLY =====\n")

    except Exception as e:
        print("Pipeline Failed:", e)

# ==========================================
# STEP 10 - SCHEDULING (SIMULATION)
# ==========================================
def scheduler():
    for i in range(2): # simulate 2 runs
        print(f"\n--- Run {i+1} ---")
        pipeline()
        time.sleep(2)

# ==========================================
# RUN PIPELINE
# ==========================================
scheduler()
```
