```python
import pandas as pd
import duckdb
```

```python
data = {
    "name": ["Amit", "Riya", "John", "Amit", "Riya"],
    "product":["Laptop","Phone","Laptop", "Tablet", "Phone"],
    "sales":[1200,800,1500,300,950]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
```

```python
con = duckdb.connect()
con.execute("CREATE TABLE sales_data AS SELECT * FROM df")
```

```python
print("\n Total & Average Sales per Product:")
query_stats = """
    SELECT
        product,
        SUM(sales) AS total_sales,
        AVG(sales) AS average_sales
    FROM sales_data
    GROUP BY product
    ORDER BY total_sales DESC
"""

# Using .df() instead of .fetchall() formats the output nicely as a DataFrame
print(con.execute(query_stats).df())
```

```python
# Top product
print("\n Top Product:")
query_top = """
    SELECT
        product,
        SUM(sales) AS total_sales
    FROM sales_data
    GROUP BY product
    ORDER BY total_sales DESC
    LIMIT 1
"""
print(con.execute(query_top).df())
```
