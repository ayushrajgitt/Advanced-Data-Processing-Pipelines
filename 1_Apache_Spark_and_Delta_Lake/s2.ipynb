```python
import pandas as pd

pdf = pd.read_excel(r"C:\Users\KIIT0001\Data_Engineering\day1(10th_march)\people_data.xlsx")

df = spark.createDataFrame(pdf)

df.show()
```

```python
df.filter(df.Age > 25).show()
```

```python
df.filter(df.City == "Bangalore").show()
```

```python
df.orderBy("Age").show()  #asc order automatically
```

```python
df.orderBy(df.Age.desc()).show() # desc order
```

```python
from pyspark.sql.functions import col
df2 = df.withColumn("AgeAfter10Years", col("Age") + 10)
df2.show() 
```

```python
df3 = df.withColumnRenamed("Age","StudentAge")
df3.show()
```

```python
df.drop("City").show()
```

```python
df.groupBy("City").count().show()
```

```python
from pyspark.sql.functions import avg
df.groupBy("City").agg(avg("Age")).show()
```

```python
from pyspark.sql.functions import max
df.select(max("Age")).show()
```

```python
df.createOrReplaceTempView("students")  #df => students
```

```python
spark.sql("SELECT * FROM students WHERE Age > 25").show()
```

```python
spark.sql("SELECT Name FROM students WHERE City='Bangalore'").show()
```

```python
spark.sql("SELECT City, COUNT(*) FROM students GROUP BY City").show()  # column count(1) as it is counting each row
```

```python
filtered_df = df.filter(df.Age > 25).toPandas()
filtered_df.to_excel("filtered_students.xlsx", index=False)
```
