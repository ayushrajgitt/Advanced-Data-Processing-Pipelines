```python
#computing some large process using pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("compute-demo").getOrCreate()
data = [
    (1,"Amit",200),
    (2,"Riya",300),
    (3,"John",150)
]
spark_df = spark.createDataFrame(data,["id","name","sales"])
spark_df.show()
```

```python
spark_df.groupBy("name").sum("sales").show()
#similar to Dataproc/EMR jobs in cloud
```
