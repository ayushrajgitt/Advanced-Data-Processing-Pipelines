```python
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .appName("SparkTask") \
    .getOrCreate()
```

```python
data = [
    ("Amit",10000),
    ("Bob",8500),
    ("Charlie",30000),
    ("Devesh",15600),
    ("Ellina",5000)
]
columns =["Name", "Salary"]
df = spark.createDataFrame(data,columns)
df.show()
```

```python
df_copy = df
df_copy.show()
```

```python
from pyspark.sql.functions import max, min,avg
df.select(
    max("Salary").alias("Max Salary"),
    min("Salary").alias("Min Salary"),
    avg("Salary").alias("Avg Salary")
).show()
```
