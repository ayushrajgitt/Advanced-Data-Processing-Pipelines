```python
from pyspark.sql import SparkSession
from delta import configure_spark_with_delta_pip

builder = SparkSession.builder \
.appName("DeltaDemo") \
.config("spark.sql.extension", "io.delta.sql.DeltaSparkSessionExtension") \
.config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog")

spark = configure_spark_with_delta_pip(builder).getOrCreate()
```

```python
data = [
    (1,"Amit",200),
    (2,"Riya",300)
]

df = spark.createDataFrame(data,["id","name","sales"])

df.write.format("delta").mode("overwrite").save("/content/delta-table")
```

```python
spark.read.format("delta").load("/content/delta-table").show()
```
