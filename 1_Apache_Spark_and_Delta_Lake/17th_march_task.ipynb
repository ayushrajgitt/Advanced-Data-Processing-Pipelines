```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("EcommerceETL").getOrCreate()
```

```python
df = spark.read.csv("sales_.csv", header=True, inferSchema=True)   # inferSchema=True automatically figures out that 'amount' is a number
```

```python
df.show()
```

```python
# Create a temporary SQL view so we can use standard SQL queries!
df.createOrReplaceTempView("sales_data")
```

```python
print("--- Insight 1: Total Revenue by Product ---")
revenue_df = spark.sql("""
    SELECT product, SUM(amount) as total_revenue
    FROM sales_data
    GROUP BY product
    ORDER BY total_revenue DESC
""")
revenue_df.show()
```

```python
print("--- Insight 2: Top Spenders ---")
spenders_df = spark.sql("""
    SELECT name, SUM(amount) as total_spent
    FROM sales_data
    GROUP BY name
    ORDER BY total_spent DESC
""")
spenders_df.show()
```
