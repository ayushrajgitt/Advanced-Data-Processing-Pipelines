```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder.appName("SparkExample").getOrCreate()

print("Spark is running")
```

```python
data = [
("Ayush",23,"Delhi"),
("Levi",None,"Mumbai"),
("Bob",30,"Bangalore"),
("Emma",28,None),
("John",22,"Hyderabad")
]

columns = ["Name","Age","City"]
df_null = spark.createDataFrame(data,columns)
df_null.show()
```

```python
df_null.printSchema()
```

```python
df_null.dropna().show()
```

```python
df_null.fillna(0).show()
```

```python
df_null.fillna({"Age":25}).show()
```

```python
df_null.fillna({"City":"Unknown"}).show()
```

```python
data2 = [
("Ayush","A"),
("Levi","B"),
("Bob","A"),
("Emma","C"),
("John","B")
]
columns2 = ["Name","Grade"]
df_grade = spark.createDataFrame(data2,columns2)

df_grade.show()
```

```python
df_grade.join(df_grade,"Name").show()  #inner join
```

```python
df_grade.join(df_grade,"Name","left").show()  #left join
```

```python
df_grade.join(df_grade,"Name","right").show()  #right join
```

```python
df_grade.join(df_grade,"Name","outer").show()  #outer join
```
