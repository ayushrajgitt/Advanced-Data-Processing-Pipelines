```python
import pandas as pd
```

```python
sales = pd.DataFrame({
    "sales_id": [1,2,3,4],
    "product_id": [101,102,101,103],
    "customer_id": [1,2,1,3],
    "date_id": [1,1,2,2],
    "amount": [50000,20000,30000,15000],
    "quantity":[1,2,1,1]
})

sales
```

```python
product = pd.DataFrame({
    "product_id": [101,102,103],
    "product_name": ["Laptop"," Phone", "Tablet"],
    "category": ["Electronics", "Electronics", "Electronics"]
})

product
```

```python
customer = pd.DataFrame({
    "customer_id": [1, 2, 3],
    "customer_name": ["Amit", "Riya", "John"],
    "city": ["Bangalore", "Mumbai", "Delhi"]
})

customer
```

```python
date = pd.DataFrame({
    "date_id": [1, 2],
    "month": ["January", "February"],
    "year": [2024, 2024]
})

date
```

```python
df = (sales
     .merge(product, on="product_id")
     .merge(customer, on="customer_id")
     .merge(date, on="date_id"))

df[["product_name", "customer_name", "city", "month", "amount"]]
```

```python
df.groupby("product_name")["amount"].sum().reset_index()
```

```python
df.groupby("customer_name")["amount"].sum().reset_index()
```

```python
df.groupby("city")["amount"].sum().reset_index()
```

```python
df.groupby("month")["amount"].sum().reset_index()
```

```python
(df.groupby("product_name")["amount"]
   .sum()
   .reset_index()
   .sort_values(by="amount", ascending=False)
   .head(1))
```

```python
(df.groupby("customer_name")["amount"]
   .sum()
   .reset_index()
   .sort_values(by="amount", ascending=False)
   .head(1))
```

```python
df["amount"].mean()
```

```python
df.groupby("product_name")["quantity"].sum().reset_index()
```
