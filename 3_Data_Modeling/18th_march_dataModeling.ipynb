```python
import pandas as pd

# Fact Table: Contains measurable data and keys to dimension tables
sales = pd.DataFrame({
    "sales_id": [1, 2, 3, 4],
    "product_id": [101, 102, 101, 103],
    "customer_id": [1, 2, 1, 3],
    "date_id": [1, 1, 2, 2],
    "amount": [50000, 20000, 30000, 15000],
    "quantity": [1, 2, 1, 1]
})

# Product Dimension
product = pd.DataFrame({
    "product_id": [101, 102, 103],
    "product_name": ["Laptop", "Phone", "Tablet"],
    "category": ["Electronics", "Electronics", "Electronics"]
})

# Customer Dimension
customer = pd.DataFrame({
    "customer_id": [1, 2, 3],
    "customer_name": ["Amit", "Riya", "John"],
    "city": ["Bangalore", "Mumbai", "Delhi"]
})

# Date Dimension
date = pd.DataFrame({
    "date_id": [1, 2],
    "month": ["Jan", "Feb"],
    "year": [2024, 2024]
})
```

```python
# Merging the Fact table with Product and Customer Dimensions
df = sales.merge(product, on="product_id") \
          .merge(customer, on="customer_id") \
          .merge(date, on="date_id")

df
```

```python
final_summary = df.groupby("product_name")["amount"].sum()
print(final_summary)
```

```python
# 2. Total sales amount per Customer
df.groupby("customer_name")["amount"].sum()
```

```python
# 3. Total sales amount per Month
df.groupby("month")["amount"].sum()
```

```python
# Sorting by amount in descending order and taking the top 1 record
df.sort_values("amount", ascending=False).head(1)
```
