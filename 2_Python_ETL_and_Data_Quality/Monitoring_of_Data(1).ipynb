```python
# ==========================================
# STEP 1 - IMPORTS
# ==========================================
import pandas as pd
import time
import logging
```

```python
# ==========================================
# STEP 2 - SETUP LOGGING
# ==========================================
logging.basicConfig(level=logging.INFO, format="%(levelname)s - %(message)s", force=True)
```

```python
# ==========================================
# STEP 3 - LOAD DATA
# ==========================================
def load_data():
    logging.info("STEP 1: Loading data...")

    df = pd.DataFrame({
        "id": [1,2,3,4],
        "amount": [100,200,"error",400]
    })

    print("\n📊 Raw Data:\n", df)
    return df
```

```python
# ==========================================
# STEP 4 - CLEAN DATA
# ==========================================
def clean_data(df):
    logging.info("STEP 2: Cleaning data...")

    try:
        df["amount"] = pd.to_numeric(df["amount"])
    except Exception as e:
        logging.error(f"Conversion failed: {e}")
        df["amount"] = pd.to_numeric(df["amount"], errors="coerce")

    if df["amount"].isnull().sum() > 0:
        logging.warning("Null values found")
        df["amount"] = df["amount"].fillna(0)

    print("\n🧹 Cleaned Data:\n", df)
    return df
```

```python
# ==========================================
# STEP 5 - TRANSFORMATION
# ==========================================
def transform_data(df):
    logging.info("STEP 3: Transforming data...")

    df["result"] = df["amount"] * 2

    print("\n🧮 Transformed Data:\n", df)
    return df
```

```python
# ==========================================
# STEP 6 - MONITOR METRICS
# ==========================================
def monitor_metrics(df):
    logging.info("STEP 4: Monitoring metrics...")

    print("\n📈 Metrics:")
    print("Row count:", len(df))
    print("Mean amount:", df["amount"].mean())
    print("Max amount:", df["amount"].max())
```

```python
# ==========================================
# STEP 7 - SAVE DATA
# ==========================================
def save_data(df):
    logging.info("STEP 5: Saving data...")

    df.to_csv("final_output.csv", index=False)

    print("\n✅ Data saved as final_output.csv")
```

```python
# ==========================================
# STEP 8 - MAIN PIPELINE
# ==========================================
def pipeline():
    logging.info("🚀 PIPELINE STARTED")

    start_time = time.time()

    try:
        df = load_data()
        df = clean_data(df)
        df = transform_data(df)
        monitor_metrics(df)
        save_data(df)

    except Exception as e:
        logging.critical(f"Pipeline failed: {e}")

    end_time = time.time()

    print(f"\n⏱️ Total Execution Time: {end_time - start_time:.4f} seconds")
    logging.info("✅ PIPELINE COMPLETED")

# ==========================================
# RUN PIPELINE
# ==========================================
pipeline()
```
