```python
import pandas as pd
df= pd.DataFrame({
"order_id": [1,2,2,4,5],
"customer_name": ["Amit", None, "Riya", "John", "Sara"],
"email": ["amit@gmail.com","riya@gmail", None, "john@gmail.com","sara@gmail.com"],
"product": ["Laptop", "Phone", "Phone", "Tablet", None],
"amount": [50000,-20000, 20000, None, 30000],
"date": ["2024-01-01","2024-01-02", "wrong_date", "2024-01-04", "2024/01/05"]
})
df
```

```python
df.isnull().sum()
```

```python
df[df.duplicated()]
```

```python
df[df["amount"]<0]
```

```python
df[~df["email"].str.contains("@", na=False)]
```

```python
pd.to_datetime(df["date"],errors='coerce')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["customer_name"].fillna("Unknown", inplace=True)
df["product"].fillna("Unknown", inplace=True)
```

```python
df["amount"] = df["amount"].apply(lambda x: abs(x) if pd.notnull(x) and x < 0 else x)
```

```python
df["date"] = pd.to_datetime(df["date"], errors='coerce')
```

```python
df["email"] = df["email"].apply(lambda x: x if "@" in str(x) else "invalid@email.com")
```

```python
df = df[df["amount"].notnull()]
```
