````python
import pandas as pd
import time

n = 150000

df = pd.DataFrame({
    "order_id": list(range(n)),
    "amount": [x % 1000 for x in range(n)],   # repeating values
    "category": ["A","B","C","A","B"] * (n // 5),
    "temp1": [x % 50 for x in range(n)],
    "temp2": [x % 20 for x in range(n)]
})

# add duplicates
df = pd.concat([df, df.iloc[:10000]], ignore_index=True)

print("Shape:", df.shape)
````

````python
start = time.time()

result = []
for x in df["amount"]:
    result.append(x * 2)

df["result_loop"] = result

end = time.time()

baseline_time = end - start

print("Time taken (baseline):", baseline_time)
print("Dataset size:", df.shape)
````

````python
start = time.time()

df["result_vectorized"] = df["amount"] * 2

end = time.time()

vector_time = end - start

print("Vectorized time:", vector_time)
````

Why vectorization is faster?

Because it uses optimized NumPy (C-level execution) instead of Python loops, processing data in bulk.

````python
before = df.shape[0]

df_clean = df.drop_duplicates().copy()

after = df_clean.shape[0]

print("Rows before:", before)
print("Rows after:", after)
````

How duplicates increase cost?

Duplicates increase data size, leading to unnecessary computation, higher memory usage, and increased processing cost.

````python
df_filtered = df_clean[df_clean["amount"] > 500].copy()

print("Filtered shape:", df_filtered.shape)
````

Reduces rows → less computation

````python
df_filtered = df_filtered[["amount"]]
````

Removes unnecessary columns → saves memory

````python
df_filtered["amount"] = df_filtered["amount"].astype("int32")
````

````python
df_filtered["amount"] = df_filtered["amount"].astype("int32")
````

````python
cache = {}

def compute(x):
    if x in cache:
        return cache[x]

    result = x * 2
    cache[x] = result
    return result

df_filtered["cached_result"] = df_filtered["amount"].apply(compute)
````

````python
def optimized_pipeline():
    start = time.time()

    df = pd.DataFrame({
        "order_id": list(range(150000)),
        "amount": [x % 1000 for x in range(150000)],
        "category": ["A","B","C","A","B"] * 30000,
        "temp1": [x % 50 for x in range(150000)],
        "temp2": [x % 20 for x in range(150000)]
    })

    # duplicates
    df = pd.concat([df, df.iloc[:10000]], ignore_index=True)

    # cleaning
    df = df.drop_duplicates()

    # filtering
    df = df[df["amount"] > 500].copy()

    # column pruning
    df = df[["amount"]]

    # memory optimization
    df["amount"] = df["amount"].astype("int32")

    # vectorized computation
    df["result"] = df["amount"] * 2

    df.to_csv("optimized_output.csv", index=False)
    print("File saved: optimized_output.csv")

    end = time.time()

    print("Final Time:", end - start)
    print("Final Shape:", df.shape)

    df.info()

    return df

final_df = optimized_pipeline()
````
