```python
import pandas as pd
import logging
import time

logging.basicConfig(
    level=logging.INFO,
    format="%(levelname)s - %(message)s",
    force=True   # important for Colab
)
```

```python
def load_data():
    logging.info(" Loading data...")

    df = pd.DataFrame({
        "id": [1,2,3,4,5],
        "amount": [100,200,"error",300,400]
    })

    print("\n Raw Data:\n", df)
    return df
```

```python
def clean_data(df):
    logging.info(" Cleaning data...")

    try:
        df["amount"] = pd.to_numeric(df["amount"])
    except Exception as e:
        logging.error(f"Conversion failed: {e}")
        df["amount"] = pd.to_numeric(df["amount"], errors="coerce")

    if df["amount"].isnull().sum() > 0:
        logging.warning("Null values found")
        df["amount"] = df["amount"].fillna(0)

    print("\n Cleaned Data:\n", df)
    return df
```

```python
def transform_data(df):
    logging.info(" Transforming data...")

    df["result"] = df["amount"] * 2

    print("\n Transformed Data:\n", df)
    return df
```

```python
def monitor_metrics(df):
    logging.info(" Monitoring metrics...")

    print("\n Metrics:")
    print("Total rows:", len(df))
    print("Average amount:", df["amount"].mean())
    print("Max amount:", df["amount"].max())
```

```python
def save_data(df):
    logging.info(" Saving data...")

    df.to_csv("output.csv", index=False)

    print("\n Data saved as output.csv")
```

```python
logging.info(" Monitoring metrics")

print("\n Metrics:")
print("Total rows:", len(df))
print("Average amount:", df["amount"].mean())
print("Max amount:", df["amount"].max())
```

```python
logging.info("Saving data")

df.to_csv("output.csv", index=False)

print("\n Data saved as output.csv")
```

```python
def pipeline():
    logging.info(" PIPELINE STARTED")

    start_time = time.time()

    try:
        df = load_data()
        df = clean_data(df)
        df = transform_data(df)
        monitor_metrics(df)
        save_data(df)

    except Exception as e:
        logging.critical(f"Pipeline failed: {e}")

    end_time = time.time()

    print(f"\n Execution Time: {end_time - start_time:.4f} seconds")
    logging.info(" PIPELINE COMPLETED")
```

```python
pipeline()
```
