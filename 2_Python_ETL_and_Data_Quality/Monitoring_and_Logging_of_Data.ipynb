```python
import pandas as pd

def load_data():
    print("[LOG] Loading data...")

    df = pd.DataFrame({
        "amount":[100,200,-50,300]
    })

    return df

df = load_data()
```

```python
def pipeline():
    print("[LOG] Pipeline started")

    df = load_data()

    print("[LOG] Data loaded")

    df["amount"] = df["amount"].apply(lambda x: abs(x))

    print("[LOG] Data cleaned")

    print("[LOG] Pipeline completed")

pipeline()
```

```python
def pipeline_with_error():
    print("[LOG] Pipeline started")

    df = pd.DataFrame({
        "amount":[100,200,"error",300]
    })

    try:
        df["amount"] = df["amount"].astype(int)
    except Exception as e:
        print("[ERROR]", e)

    print("[LOG] Pipeline finished")

pipeline_with_error()
```
