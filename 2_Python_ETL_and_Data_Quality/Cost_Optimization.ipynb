```python
# ==========================================
# STEP 1 - IMPORTS
# ==========================================
import pandas as pd
import time

print("\n===== COST OPTIMIZATION DEMO =====\n")

# ==========================================
# STEP 2 - CREATE LARGE DATA (RAW DATA)
# ==========================================
print("[STEP] Creating large dataset...\n")

df = pd.DataFrame({
    "amount": list(range(100000)),
    "category": ["A", "B", "C", "A", "B"] * 20000
})

# Add duplicates intentionally
df = pd.concat([df, df])

print("Data created. Shape:", df.shape)

# ==========================================
# STEP 3 - BAD APPROACH (COSTLY)
# ==========================================
print("\n[STEP] Running COSTLY pipeline...\n")

start = time.time()

# Process full data using loop
result = []
for x in df["amount"]:
    result.append(x * 2)

df["result"] = result

# No filtering
# No cleaning

end = time.time()

print("Time taken (COSTLY):", end - start)

# ==========================================
# STEP 4 - OPTIMIZED APPROACH
# ==========================================
print("\n[STEP] Running OPTIMIZED pipeline...\n")

start = time.time()

# 1. Remove duplicates
df_opt = df.drop_duplicates().copy()

# 2. Filter only required data
df_opt = df_opt[df_opt["amount"] > 50000]

# 3. Select only required columns
df_opt = df_opt[["amount"]]

# 4. Optimize datatype
df_opt["amount"] = df_opt["amount"].astype("int32")

# 5. Use vectorized operation (fast)
df_opt["result"] = df_opt["amount"] * 2

end = time.time()

print("Time taken (OPTIMIZED):", end - start)

# ==========================================
# STEP 5 - CACHING (SAVE COMPUTE)
# ==========================================
print("\n[STEP] Applying caching...\n")

cache = {}

def compute(x):
    if x in cache:
        return cache[x]

    result = x * 2
    cache[x] = result
    return result

start = time.time()

df_opt["cached_result"] = df_opt["amount"].apply(compute)

end = time.time()

print("Time with caching:", end - start)

# ==========================================
# STEP 6 - MEMORY USAGE
# ==========================================
print("\n[STEP] Memory usage (optimized):\n")
df_opt.info()

# ==========================================
# STEP 7 - SAVE OPTIMIZED DATA
# ==========================================
df_opt.to_csv("optimized_data.csv", index=False)

print("\nOptimized data saved successfully!")

# ==========================================
# FINAL SUMMARY
# ==========================================
print("\n===== SUMMARY =====")
print("✓ Reduced data size")
print("✓ Removed duplicates")
print("✓ Filtered unnecessary data")
print("✓ Used vectorization instead of loop")
print("✓ Applied caching")
print("✓ Reduced memory usage")
```
