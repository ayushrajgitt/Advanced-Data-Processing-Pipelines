```python
import pandas as pd

n = 100000

df = pd.DataFrame({
    "order_id": list(range(n)),
    "amount": [x % 1000 for x in range(n)],   # repeating values (for caching)
    "category": ["A" if x % 2 == 0 else "B" for x in range(n)]
})

# add duplicates
df = pd.concat([df, df.iloc[:5000]], ignore_index=True)

df.head()
```

```python
import time

start = time.time()

result = []
for x in df["amount"]:
    result.append(x * 2)

df["result_loop"] = result

loop_time = time.time() - start
print("Loop Time:", loop_time)
```

```python
start = time.time()

df["result_vectorized"] = df["amount"] * 2

vector_time = time.time() - start
print("Vectorized Time:", vector_time)
```

```python
print("Loop Time:", loop_time)
print("Vectorized Time:", vector_time)
```

Vectorized is faster because it uses optimized C-level operations (NumPy) instead of Python loops.

```python
df["amount"] = df["amount"].astype("int32")
df["result_vectorized"] = df["result_vectorized"].astype("int32")

df.info()
```

```python
cache = {}

def compute(x):
    if x in cache:
        print(f"Using cache for {x}")
        return cache[x]

    print(f"Computing for {x}")
    result = x * 2
    cache[x] = result
    return result

# small sample to visualize
sample = pd.DataFrame({"amount": [10,20,10,30,20]})
sample["result"] = sample["amount"].apply(compute)

sample
```

```python
def optimized_pipeline():
    start = time.time()

    # Load
    df = pd.DataFrame({
        "order_id": list(range(100000)),
        "amount": [x % 1000 for x in range(100000)],
        "category": ["A" if x % 2 == 0 else "B" for x in range(100000)]
    })

    # Add duplicates
    df = pd.concat([df, df.iloc[:5000]], ignore_index=True)

    # Remove duplicates
    df = df.drop_duplicates()

    # Filtering
    df = df[df["amount"] > 100].copy()

    # Vectorized computation
    df["result"] = df["amount"] * 2

    # Memory optimization
    df["amount"] = df["amount"].astype("int32")
    df["result"] = df["result"].astype("int32")

    end = time.time()

    print("Final Time:", end - start)
    print("\nFinal Data:")
    print(df.head())

    print("\nMemory Usage:")
    df.info()

    return df

optimized_df = optimized_pipeline()
```
