```python
import pandas as pd
import time

df = pd.DataFrame({
    "amount": list(range(100000))
})

start = time.time()

result = []
for x in df["amount"]:
    result.append(x * 2)

df["result"] = result

end = time.time()

print("Time taken (loop):", end - start)
```

```python
cache = {}

def compute(x):
    if x in cache:
        print(f"Using cache for {x}")
        return cache[x]

    print(f"Computing for {x}")
    result = x * 2
    cache[x] = result
    return result

df = pd.DataFrame({
    "amount": [10, 20, 10, 30, 20]
})

df["result"] = df["amount"].apply(compute)

df
```
