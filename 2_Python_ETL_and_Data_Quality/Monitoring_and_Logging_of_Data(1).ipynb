```python
# ==========================================
# STEP 1 - IMPORTS
# ==========================================
import pandas as pd
import logging
import time
```

```python
# ==========================================
# STEP 2 - SETUP LOGGING
# ==========================================
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    filename="pipeline.log", # logs saved to file
    filemode="w" # overwrite each run
)

# Also print logs to console
console = logging.StreamHandler()
console.setLevel(logging.INFO)
formatter = logging.Formatter("%(levelname)s - %(message)s")
console.setFormatter(formatter)
logging.getLogger().addHandler(console)
```

```python
# ==========================================
# STEP 3 - CREATE DATA
# ==========================================
def load_data():
    logging.info("Loading data...")

    df = pd.DataFrame({
        "id": [1,2,3,4],
        "amount": [100,200,"error",400]
    })

    logging.info(f"Data loaded with {len(df)} rows")
    return df
```

```python
# ==========================================
# STEP 4 - DATA CLEANING
# ==========================================
def clean_data(df):
    logging.info("Cleaning data...")

    try:
        df["amount"] = pd.to_numeric(df["amount"])
    except Exception as e:
        logging.error(f"Data type conversion failed: {e}")
        df["amount"] = pd.to_numeric(df["amount"], errors="coerce")

    # Handle missing values
    if df["amount"].isnull().sum() > 0:
        logging.warning("Null values found in 'amount'")
        #df["amount"].fillna(0, inplace=True)
        df["amount"] = df["amount"].fillna(0)

    logging.info("Data cleaning completed")
    return df
```

```python
# ==========================================
# STEP 5 - TRANSFORMATION
# ==========================================
def transform_data(df):
    logging.info("Starting transformation...")

    df["result"] = df["amount"] * 2

    logging.info("Transformation completed")
    return df
```

```python
# ==========================================
# STEP 6 - SAVE DATA
# ==========================================
def save_data(df):
    logging.info("Saving data...")

    df.to_csv("output.csv", index=False)

    logging.info("Data saved successfully")
```

```python
# ==========================================
# STEP 7 - MAIN PIPELINE
# ==========================================
def pipeline():
    logging.info("========== PIPELINE STARTED ==========")

    start_time = time.time()

    try:
        df = load_data()
        df = clean_data(df)
        df = transform_data(df)
        save_data(df)

    except Exception as e:
        logging.critical(f"Pipeline failed: {e}")

    end_time = time.time()

    logging.info(f"Pipeline completed in {end_time - start_time:.4f} seconds")
    logging.info("========== PIPELINE ENDED ==========")
```

```python
# ==========================================
# RUN PIPELINE
# ==========================================
pipeline()
```
