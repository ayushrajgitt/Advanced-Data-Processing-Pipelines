```python
import pandas as pd

df = pd.DataFrame({
    "order_id":[1,2,2,4,5,6],
    "customer_name":["Amit",None,"Riya","John","Sara","Amit"],
    "email":["amit@gmail.com","riya@gmail","riya@gmail.com","john@gmail.com","sara@gmail","amit@gmail.com"],
    "product":["Laptop","Phone","Phone","Tablet",None,"Laptop"],
    "amount":[50000,-20000,20000,None,30000,50000],
    "quantity":[1,2,2,1,None,1],
    "date":["2024-01-01","2024-01-02","wrong_date","2024-01-04","2024/01/05","2024-01-01"]
})

df
```

```python
df.isnull().sum()
```

```python
df[df.duplicated()]
```

```python
df[df["amount"] < 0]
```

```python
df[~df["email"].str.contains("@", na=False)]
```

```python
pd.to_datetime(df["date"], errors='coerce')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["customer_name"] = df["customer_name"].fillna("Unknown")
df["product"] = df["product"].fillna("Unknown")
```

```python
df["amount"] = df["amount"].apply(lambda x: abs(x) if pd.notnull(x) and x < 0 else x)
```

```python
df["email"] = df["email"].apply(
    lambda x: x if pd.notnull(x) and "@" in x and "." in x else "invalid@email.com"
)
```

```python
df["date"] = pd.to_datetime(df["date"], errors='coerce')
```

```python
df = df[df["date"].notnull()]
```

```python
df = df[df["amount"].notnull() & df["quantity"].notnull()]
```

```python
print("Null values:\n", df.isnull().sum())
print("\nDuplicates:\n", df.duplicated().sum())
print("\nAny negative amounts:\n", (df["amount"] < 0).sum())
print("\nDates:\n", df["date"])
```
