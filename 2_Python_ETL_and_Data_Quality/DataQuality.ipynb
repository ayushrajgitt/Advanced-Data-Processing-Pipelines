```python
import pandas as pd # create data

df = pd.DataFrame({
    "id":[1,2,2,4],
    "name":["Amit",None,"Riya","John"],
    "sales":[200,-50,300,None]
})

df
```

```python
df.isnull().sum()  #null check
```

```python
df.duplicated() # duplicate check
```

```python
df[df["sales"]<0] #range check
```

```python
df["name"].fillna("Unknown", inplace=True) fix datas [it should be an error]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["sales"] = df["sales"].apply(lambda x: abs(x) if x < 0 else x)
```
