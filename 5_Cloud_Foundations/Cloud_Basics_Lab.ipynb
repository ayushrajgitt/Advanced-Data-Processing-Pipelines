```python
import pandas as pd

data = {
    "id":[1,2,3,4,5],
    "name":["Amit","Riya","John","Sara","David"],
    "sales":[200,300,150,400,250]
}

df = pd.DataFrame(data)

df
```

```python
df["sales"].mean()
```

```python
df.groupby("name")["sales"].sum()
```

```python
import matplotlib.pyplot as plt

df.groupby("name")["sales"].sum().plot(kind="bar")
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("cloud_demo").getOrCreate()

data =[
    (1,"Amit",200),
    (2,"Riya",300),
    (3,"John",500)
]

spark_df = spark.createDataFrame(data,["id","name","sales"])

spark_df.show()
```

```python
spark_df.groupBy("name").sum("sales").show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "id":[1,2,3],
    "product":["phone","laptop","camera"],
    "price":[20000,60000,80000]
}

df = pd.DataFrame(data)

df
```

```python
df["price"].mean()
```

```python
df.groupby("product")["price"].sum().plot(kind="bar")
```
